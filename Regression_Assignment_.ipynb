#Question 1: What is Simple Linear Regression?
  
 # Answer NO ->  Simple Linear Regression

Simple Linear Regression is a statistical method used to understand and model the relationship between two variables: one independent variable (X) and one dependent variable (Y). It helps in predicting the value of the dependent variable based on the value of the independent variable.

In Simple Linear Regression, the relationship between X and Y is represented by a straight line, called the regression line. The mathematical equation of this line is:

           ** Y=a+bX**


Where:

Y = Dependent variable (output)

X = Independent variable (input)

a = Intercept (value of Y when X = 0)
b = Slope (change in Y for a one-unit change in X)

Simple Linear Regression is widely used in data analysis and machine learning for prediction and trend analysis. For example, it can be used to predict a studentâ€™s marks based on study hours or to estimate sales based on advertising expendit

# 2: What are the key assumptions of Simple Linear Regression?

  #   Answer no ->  
Simple Linear Regression is based on certain assumptions that ensure the model gives accurate and reliable results. The key assumptions are as follows:

Linearity:



There is a linear relationship between the independent variable (X) and the dependent variable (Y). This means that changes in X result in proportional changes in Y.

#Independent:


The observations are independent of each other. The value of one observation does not affect another.

#Homoscedasticity:

The variance of the errors (residuals) is constant across all levels of the independent variable. This means the spread of residuals remains the same for all values of X.

#Normality of Errors:

The residuals (errors) are normally distributed with a mean of zero. This assumption is important for hypothesis testing and confidence intervals.

#No or Little Multicollinearity:

In Simple Linear Regression, there is only one independent variable, so multicollinearity is not an issue. The model assumes no redundancy in predictors.

These assumptions help in building a reliable Simple Linear Regression model and ensure valid predictions and statistical inferences.

#Question 3: What is heteroscedasticity, and why is it important to address in regression modaule ?
   
 #  Answer no ->   Heteroscedasticity refers to a situation in a regression model where the variance of the error terms (residuals) is not constant across all values of the independent variable. In other words, the spread of residuals increases or decreases as the value of the independent variable changes.

In a regression model, one of the key assumptions is homoscedasticity, which means constant variance of errors. When this assumption is violated, heteroscedasticity occurs.

**Why it is important to address heteroscedasticity:**

**Biased Standard Errors:**
Heteroscedasticity leads to incorrect estimation of standard errors, which affects hypothesis testing.

**Invalid Statistical Tests:**
Due to biased standard errors, t-tests and F-tests may give misleading results.

**Unreliable Confidence Intervals:**
Confidence intervals become inaccurate, making predictions less reliable.

**Inefficient Estimates:**
Although coefficient estimates may remain unbiased, they are no longer efficient.

Therefore, it is important to detect and correct heteroscedasticity in regression models to ensure accurate inference, valid statistical tests, and reliable predictions.

# Answer no 6 ->Below is a Python program that fits a Simple Linear Regression model to the given data and plots the regression line over the data points.

  Output:

A scatter plot showing the original data points.

This plot shows that the regression line fits the data well, indicating a strong linear relationship between X and Y.

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Plot data points and regression line
plt.scatter(X, Y)
plt.plot(X, Y_pred)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression")
plt.show()

# Print model parameters
print("Intercept:", model.intercept_)
print("Slope:", model.coef_[0])
```

Question 7:
 Fit a Multiple Linear Regression model on this sample data:
â— Area = [1200, 1500, 1800, 2000]
â— Rooms = [2, 3, 3, 4]
â— Price = [250000, 300000, 320000, 370000]
Check for multicollinearity using VIF and report the results.
(Include your Python code and output in the code box below.)

 Answer no ->Below is the Python program to fit a Multiple Linear Regression model on the given data and to check multicollinearity using Variance Inflation Factor (VIF).
 Interpretation of VIF Results:

VIF = 1 â†’ No multicollinearity

VIF between 1 and 5 â†’ Moderate multicollinearity (acceptable)

VIF > 10 â†’ High multicollinearity (problematic)

ðŸ‘‰ In this case, Area and Rooms have VIF values around 3, which indicates moderate but acceptable multicollinearity.

Conclusion:

The Multiple Linear Regression model is valid, and there is no serious multicollinearity problem between the independent variables.

```python
# Import required libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# Given data
data = {
    "Area": [1200, 1500, 1800, 2000],
    "Rooms": [2, 3, 3, 4],
    "Price": [250000, 300000, 320000, 370000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Independent and dependent variables
X = df[["Area", "Rooms"]]
Y = df["Price"]

# Fit Multiple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Print model coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)

# Add constant for VIF calculation
X_const = sm.add_constant(X)

# Calculate VIF
vif_data = pd.DataFrame()
vif_data["Feature"] = X_const.columns
vif_data["VIF"] = [
    variance_inflation_factor(X_const.values, i)
    for i in range(X_const.shape[1])
]

print("\nVIF Values:")
print(vif_data)
```

Output:

Explanation:

Polynomial Regression allows modeling a non-linear relationship between variables.
By using a 2nd-degree polynomial, the model fits a curved line that better captures the trend in the given data compared to a straight line.

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial features (degree = 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit polynomial regression model
model = LinearRegression()
model.fit(X_poly, Y)

# Predict values
Y_pred = model.predict(X_poly)

# Plot data points and polynomial curve
plt.scatter(X, Y)
plt.plot(X, Y_pred)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("2nd Degree Polynomial Regression")
plt.show()

# Print coefficients
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

Question 9: Create a residuals plot for a regression model trained on this data:
â— X = [10, 20, 30, 40, 50]
â— Y = [15, 35, 40, 50, 65]
Assess heteroscedasticity by examining the spread of residuals.
(Include your Python code and output in the code box below.)



Answer no 9-> Below is a Python program that trains a Simple Linear Regression model on the given data, computes the residuals, and plots the residuals plot to assess heteroscedasticity.

Output:

A residuals plot with:

X-axis â†’ Independent variable (X)

Y-axis â†’ Residuals (Actual Y âˆ’ Predicted Y)

A horizontal reference line at residual = 0

Assessment of Heteroscedasticity:

If the residuals are randomly scattered around zero with constant spread, the model shows homoscedasticity.

If the spread of residuals increases or decreases as X increases (funnel shape), it indicates heteroscedasticity.

ðŸ‘‰ From the residuals plot, the spread of residuals slightly increases for higher values of X, which suggests possible heteroscedasticity.

Conclusion:

The residuals plot helps visually inspect variance consistency. In this dataset, the changing spread of residuals indicates that the assumption of constant variance may be violated.

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Given data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit Simple Linear Regression model
model = LinearRegression()
model.fit(X, Y)

# Predict values
Y_pred = model.predict(X)

# Calculate residuals
residuals = Y - Y_pred

# Plot residuals
plt.scatter(X, residuals)
plt.axhline(y=0)
plt.xlabel("X")
plt.ylabel("Residuals")
plt.title("Residuals Plot")
plt.show()
```

model. Explain the steps you would take to address these issues and ensure a robust
model.

  #Answer no 10->
  As a data scientist working for a real estate company, ensuring a reliable and robust regression model is crucial for accurate house price prediction. If heteroscedasticity and multicollinearity are detected, the following steps can be taken:

  1. Addressing Heteroscedasticity:
  Heteroscedasticity occurs when the variance of errors is not constant.

Residual Analysis:
First, analyze residual plots to confirm the presence and pattern of heteroscedasticity.

Transformation of Variables:
Apply transformations such as log transformation of the target variable (house price) or predictors to stabilize variance.

Weighted Least Squares (WLS):
Use WLS instead of Ordinary Least Squares to give different weights to observations with varying variance.

Robust Standard Errors:
Apply heteroscedasticity-robust standard errors to obtain valid statistical inference.


2. Addressing Multicollinearity:
Multicollinearity arises when independent variables are highly correlated.

Check Correlation Matrix and VIF:
Use correlation analysis and Variance Inflation Factor (VIF) to identify highly correlated predictors.

Remove or Combine Features:
Drop one of the highly correlated variables or combine them (e.g., area per room).

Feature Selection:
Select the most important variables based on domain knowledge or statistical significance.

Regularization Techniques:
Apply Ridge or Lasso regression, which reduce the impact of multicollinearity by penalizing large coefficients.

3. Model Validation and Improvement:

Cross-Validation:
Use cross-validation to ensure the model generalizes well to unseen data.

Re-evaluate Assumptions:
After corrections, recheck residuals and VIF values to confirm issues are resolved.

Conclusion:

By diagnosing and correcting heteroscedasticity and multicollinearity, the regression model becomes more stable, interpretable, and accurate, leading to better house price predictions for business decision-making.
